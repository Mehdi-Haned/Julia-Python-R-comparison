```r
# Is_even mutual recursion

is_even <- function(n) {
    if (n == 0) {
        return(TRUE)
    } else {
        return(is_odd(n - 1))
    }
}

is_odd <- function(n) {
    if (n == 0) {
        return(FALSE)
    } else {
        return(is_even(n - 1))
    }
}

is_even(1001)
```

```r
# First N fibboncci numbers

fib <- function(n) {
    ans <- 1:n
    ans[2] <- 1
    for (i in  3:n) {
        ans[i] <- ans[i - 1] + ans[i - 2]
    }
    return(ans)
}

fib(100)
```

```r
# is it a prime

is_prime <- function(n) {
    for (i in 2:(n - 1)) {
        if (all(c(i > 1, n %% i == 0, i != n))) {
            return(FALSE)
        }
    }
    return(TRUE)
}

is_prime(2^13 - 1)
```

```r
# first n primes

first_n_primes <- function(n) {
    res <- c()
    for (i in 2:n) {
        if (is_prime(i)) {
            res <- append(res, i)
        }
    }
    return(res)
}

first_n_primes(100)
```
